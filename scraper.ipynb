```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd 
```

```python
 # get link 
url = f"https://www.premierleague.com/tables"

# get content
website = requests.get(url).content

# use beautifulsoup to parse data
soup = BeautifulSoup(website, 'lxml')
```

```python
premier_league = soup.find(class_='tableBodyContainer isPL')
team_name = premier_league.find_all(class_='team')
list_of_teams = []
for team in team_name:
    list_of_teams.append(team.text.strip().split('\n')[0])
list_of_teams
```

# Top Scorers

```python
 # get link 
url = f"https://analytics.soccerment.com/en/league/premier_league/players/stat/Goals"
# get content
website = requests.get(url).content
# use beautifulsoup to parse data
soup = BeautifulSoup(website, 'lxml')


# card for top players
card = soup.find(class_='card mb10')

# lists of players names and stat
all_players_name_goals = []
all_players_goals = []

# top scorer in the top scorer card
top_scorer_name = card.find(class_='card-subtitle mb-2 player_name').text
all_players_name_goals.append(top_scorer_name)
top_scorer_goals = card.find(class_='player-value').text
all_players_goals.append(top_scorer_goals)

# rest of players in top scorer card
rest_of_players_name = card.find_all(class_='c_name')
for player in rest_of_players_name:
    all_players_name_goals.append(player.text.strip().split('|')[0])

rest_of_players_goals = card.find_all(class_='c_value')
for player in rest_of_players_goals:
    all_players_goals.append(player.text)
```
